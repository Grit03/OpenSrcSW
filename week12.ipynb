```python
import csv
f=open('Seoul.csv', encoding='cp949')
data = csv.reader(f)
header = next(data)
mx_temperatureRange= 0.0
minTemperature = 0.0
maxTemperature = 0.0
label =['']*3

for row in data:  
    temperatureRange = float(row[4]) - float(row[3])
    dayMinTemp = float(row[3])
    dayMaxTemp = float(row[4])
    if mx_temperatureRange < temperatureRange:
        mx_temperatureRange = temperatureRange
        label[0] = row[0]
        
    if minTemperature> dayMinTemp :
        minTemperature = dayMinTemp
        label[1] = row[0]
        
    if maxTemperature<dayMaxTemp :
        maxTemperature = dayMaxTemp
        label[2] = row[0]
        
print("일교차가 가장 큰 날")
print("날짜 " + label[0] + "\n")

print("최저기온이 가장 낮은 날")
print("날짜 " + label[1]+ "\n")

print("최고기온이 가장 높은 날")
print("날짜 " + label[2])

f.close();
```

```python
import csv
f=open('subway.csv')
data = csv.reader(f)
header = next(data)

subwayDic = {}
for row in data:
    if row[1] not in subwayDic:
        subwayDic[row[1]] = []

f=open('subway.csv')
data = csv.reader(f)
header = next(data)

for row in data:
    subwayNum = row[1]
    stationName = row[3]
    client = int(row[-1]) + int(row[-2])
    subwayDic[subwayNum].append((stationName, client))

for subwayNum in subwayDic:
    # 정렬
    subwayDic[subwayNum].sort(key=lambda x: x[1],reverse=True)
    data = subwayDic[subwayNum]
    # 출력할 부분 슬라이싱
    dataTop = data[0:3]
    dataBottom = data[-1:-4:-1]
    print("========== "+subwayNum+" ==========")
    print("이용객이 가장 많은 역 순위")
    for i, info in enumerate(dataTop):
        print("{0:d}. {1} ({2:,d}명)".format(i+1,info[0],info[1]))
        
    print("\n이용객이 가장 적은 역 순위")
    for i, info in enumerate(dataBottom):
        print("{0:d}. {1} ({2:,d}명)".format(i+1,info[0],info[1]))
    print()
```
